```python
import matplotlib.pyplot as plt
import numpy as np
#creating data points
x=np.linspace(0,100,30)
y=np.random.random(x.shape)

#prepare figure
plt.figure(figsize=(4,3))
ax = plt.axes()
ax.scatter(x,y)#plot data point as in scatter plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#creating data points
x=np.linspace(0,20,10)
e=np.random.random(x.shape)
y=0.5*x+1+e
#prepare figure
plt.figure(figsize=(4,3))
ax = plt.axes()
#plot data in liner data point
ax.scatter(x,y)
ax.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#create data points
x = np.linspace(0,2*np.pi,100)
y = np.cos(x)

#prepare figure
plt.figure(figsize=(3,4))
ax = plt.axes()

ax.plot(x,y)
ax.plot(x,np.zeros(x.shape),'--')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#create data points
x = np.linspace(0.001, 10, 100)
y = np.log(x)

#prepare figure
plt.figure(figsize=(6,3))
ax = plt.axes()

ax.plot(x,y)
ax.plot(x,np.zeros(x.shape),'--')

plt.show()
```
